# Data Representation

Represent the following data in stem-and-leaf plot, a histogram and a boxplot 

1. 20 21 20 19 21 19

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1  19 21 19 20 19 20 21 20  lenght of nails 
a_1 = np.array([20, 21, 20, 19, 21, 19])
n,bins , patches = plt.hist(a_1, bins=3)
plt.xlabel('Length of Nails')
plt.ylabel('No of people')
plt.grid(True)
plt.show()
```

```python
#2    6 6 4 2 1 7 0 4 6 7 calls every minute 
#minute view 
calls_every_min = [6,6,4,2,1,7,0,4,6,7]
plt.bar([_ for _ in range(len(calls_every_min))], calls_every_min, color='g')
plt.show()
```

```python
#frequency view  calls ever minute 
a_1 = np.array([6, 6, 4, 2, 1, 7, 0, 4, 6, 7])
n,bins , patches = plt.hist(a_1)
plt.xlabel('Calls per minute')
plt.ylabel('calls')
plt.grid(True)
plt.show()
```
